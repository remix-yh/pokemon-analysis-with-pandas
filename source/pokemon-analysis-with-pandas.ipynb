### 読込処理

```python
import pandas as pd

df = pd.read_csv('../dataset/pokemon_status.csv', sep=',', encoding="shift-jis")
```

### CSVの先頭5行を表示すること(head()関数を使用)

```python
df.head()
```

### CSVの合計値が高い順5行を表示すること

```python
df.sort_values(by='合計', ascending=True).head()
```

### CSVの合計値が低い順5行を表示すること

```python
df.sort_values(by='合計', ascending=False).head()
```

### 列単位で欠損値の合計数を表示すること

```python
df.isnull().sum()
```

### 要約統計量を表示すること。欠損値は0として扱うこと(describe()関数を使用)

```python
df.fillna(0).describe()
```
